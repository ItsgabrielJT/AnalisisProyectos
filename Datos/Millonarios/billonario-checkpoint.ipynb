## Datos Atipicos (outliers)

- Para ver si exiten valores atipicos tenemos que ver la distribucion de los datos

```python
plot = (prueba1['demographics.age'].value_counts()).plot(
kind='bar', figsize=(120, 10), title='Rankings')
```

- Conlusion: Vemos que la grafica esta con un pico en el cero, con una cola hacia la derecha

- PAra ver de una mejor forma los valores atipicos, applicamos un boxplot o grafico de caja

```python
# Box plot
sns.boxplot(data=prueba1,
            x = prueba1['demographics.age'].value_counts())
```

- Conclusion: el grafico de caja nos deja ver os valores atipicos que estan fuera del limite superior

- Tenemos que eliminar esos valores atipicos para evitar sesgos en futuros analisis

```python
prueba1.drop(prueba1[prueba1['demographics.age'] <= 0].index, inplace=True)
```

```python
plot = (prueba1['demographics.age'].value_counts()).plot(
kind='bar', figsize=(120, 10), title='Rankings')
```

```python
# Box plot
sns.boxplot(data=prueba1,
            x = prueba1['demographics.age'].value_counts())
```

- Conclusion: POdemos ver que eso valores atipicos ya quedaron eliminados al volver a hacer las grficas
