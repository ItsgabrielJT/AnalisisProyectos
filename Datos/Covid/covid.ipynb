```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
data = pd.read_csv('./data_original/covid_table.csv')
print('Original data')
print((data.size, 4))
print(data.head(5))
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.histplot(data=data, x='Total Deaths', stat='frequency')
plt.show()
```

```python
data.isnull()
```

```python
msno.bar(data)
```

```python
data.drop(['New Cases', 'New Deaths', 'New Recovered', 'Active Cases/1M pop', 'New Deaths/1M pop'], axis=1,inplace=True)
data
```

```python
prueba1 = data.dropna()
prueba1
```

```python
sns.histplot(data=prueba1, x='Total Deaths', stat='count')
plt.show()
```

```python
sns.boxplot(data=prueba1, x='Total Deaths', dodge=True)
plt.show()
```

```python
sns.boxplot(data=prueba1, x='Total Cases', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(prueba1.corr(), annot=True)
plt.show()
```
